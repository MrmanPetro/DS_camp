## `Iris` dataset 

```python
import pandas as pd 
import numpy as np 
np.random.seed = 2021
from sklearn.datasets import load_iris
iris = load_iris()
print ('data contains:',iris.keys())
X, y, labels, feature_names  = iris.data, iris.target, iris.target_names, iris['feature_names']
df_iris= pd.DataFrame(X, columns= feature_names) 
df_iris['label'] =  y
features_dict = {k:v for k,v in  enumerate(labels)}
df_iris['label_names'] = df_iris.label.apply(lambda x: features_dict[x])
df_iris
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

k_best = 0
score_best = 0
for k in range(1, 50):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    score = knn.score(X_test, y_test)
    #print(f"k = {k}; score = {score}")
    if score > score_best:
        score_best = score
        k_best = k
        
print ('The best k = {}, score = {}'.format(k_best,score_best ))
```

## Synthetic dataset  

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
%matplotlib notebook
from matplotlib.colors import ListedColormap
cmap_bold = ListedColormap(['blue','#FFFF00','black','green'])

np.random.seed= 2021
X_D2, y_D2 = make_blobs(n_samples = 300, n_features = 2, centers = 8,
                       cluster_std = 1.3, random_state = 4)
y_D2 = y_D2 % 2
plt.figure()
plt.title('Sample binary classification problem with non-linearly separable classes')
plt.scatter(X_D2[:,0], X_D2[:,1], c=y_D2,
           marker= 'o', s=30, cmap=cmap_bold)

```

```python
from sklearn.neighbors import KNeighborsRegressor

X_train, X_test, y_train, y_test = train_test_split(X_D2, y_D2, random_state=0)

k_best = 0
score_best = 0
for k in range(1, 50):
    knr = KNeighborsRegressor(n_neighbors=k)
    knr.fit(X_train, y_train)
    score = knr.score(X_test, y_test)
    #print(f"k = {k}; score = {score}")
    if score > score_best:
        score_best = score
        k_best = k
        
print ('The best k = {}, score = {}'.format(k_best,score_best ))
```

```python
import ML_mst as mst

%matplotlib notebook

clf = KNeighborsRegressor(n_neighbors = k_best)
clf.fit(X_train, y_train)
score = clf.score (X_test, y_test)
mst.plot_decision_boundary(clf, X_train, y_train, X_test=X_test, y_test= y_test,
                               title='KNN classification K= {}, score = {:.2f}'.format(k_best, score), 
                               precision=0.01,plot_symbol_size = 50)
```
