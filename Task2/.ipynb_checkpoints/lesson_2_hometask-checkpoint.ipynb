```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
np.random.seed(2023)
a = np.random.randint(1,100,20)
a
```

```python
plt.plot (a)
```

```python
deep = 0
bottom = 0
max_deep = 0
top1_max = 0
top2_max = 0
top1_max_index = 0
top2_max_index = 0


for i in range(len(a) - 1):
    
    # find bottom
    if a[i] > a[i+1]:
        while a[i] > a[i+1]:
            bottom = a[i+1]
            bottom_index = i+1
            i += 1
    else:
        bottom = a[0]
        bottom_index = 0
    
    # find top 1
    bottom_index1 = bottom_index 
    while a[bottom_index1] < a[bottom_index1-1]:
        if bottom_index1 > 0:
            top1 = a[bottom_index1-1]
            top1_index = bottom_index1-1
            bottom_index1 -= 1
        else:
            top1 = a[0]
            break
    
    # find top 2
    bottom_index2 = bottom_index
    while a[bottom_index2] < a[bottom_index2+1]:
        if bottom_index2 < len(a) - 2:
            top2 = a[bottom_index2 + 1]
            top2_index = bottom_index2 + 1
            bottom_index2 += 1
        else:
            top2 = a[-1]
            break
    
    # find deep
    if top1 > top2:
        deep = top2 - bottom
    else:
        deep = top1 - bottom
        
    # find max deep    
    if max_deep < deep:
        max_deep = deep
        top1_max = top1
        top2_max = top2
        top1_max_index = top1_index
        top2_max_index = top2_index

        
print("max deep:", max_deep)
print("top1 max:", top1_max)
print("top2 max:", top2_max)

deepest_valley = a[top1_max_index : top2_max_index + 1]
plt.plot(a)
offset = np.where(a == deepest_valley[0])[0][0]
plt.plot(np.arange(offset, offset + len(deepest_valley)), deepest_valley, c='red')
plt.show()
```
